# Les moindres carrés

On dispose de données: des inputs `x[i]` et des outputs `y[i]`. On cherche à expliquer l'influence des premier sur les second. 

ex: 
* `x[i]` est la surface du `i`-ième appartement
* `y[i]` est le prix du `i`-ième appartement

Dans ce cas, il est naturelle de supposer qu'il y a une relation affine entre `x` et `y`: 

	y[i] = w[0] + w[1] x[i] + bruit[i]

Dans le programme suivant, on crée des données artificielles qui suivent parfaitement ce modèle. On estime `w` par la formule des moindres carrés (cf. explication plus bas).  

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
% matplotlib inline
np.set_printoptions(precision=2,suppress=True)
```

```python
def createData_one_input(nbData: int, sigma=0.3):
    x = np.random.random(nbData) * 2
    w1 = 3
    w0 = 5
    y = w1 * x + w0 + np.random.normal(0, sigma, size=[nbData])
    return x, y

nbData=100
x,y=createData_one_input(nbData)
x_ext=np.zeros(shape=[nbData,2])
x_ext[:,0]=np.ones(shape=nbData)
x_ext[:,1]=x

""" ( X^T X )^(-1) """
XTX_1 = np.linalg.inv(x_ext.T @ x_ext)
hat_w = XTX_1 @ x_ext.T @ y

plt.plot(x, y, '.')
xx=np.linspace(0,2,10)
plt.plot(xx, hat_w[0]+hat_w[1]*xx);
```

*** Question: *** Est-ce que cela marche aussi avec les données suivantes ? 

```python
def create_deterministic_input(nbData:int):
    x=np.linspace(0,3,nbData)
    y=x*np.sin(10*x)+x+1
    return x,y
x,y=create_deterministic_input(200)
plt.plot(x,y,'.');
```

*** Exo: *** 

Notons $X$ la matrice a deux colonnes. La première colonne est constante: $X_{i,0}=1$, la seconde colonne contient les input: $X_{i,1}=$`x[i]`.  Notons Y la matrice colonne contenant les output `y`. 

La technique des moindres carrés consiste à trouver 
$$
\hat w =  \hbox{argmin}_w \| Y  -  X w \|^2 =   \hbox{argmin}_w ( Y  -  X w)^T (Y -  X w)  
$$
le minimum étant cherché parmi tous les vecteurs colonne $w=(w_0,w_1) \in \mathbb{R}^2$. 

Calculez la différentelle de $w \to  ( Y  - w X)^T (Y - w X) $ en développant le produit matriciel. Cherchez l'endroit où cette différentielle s'annule. Vous devrez retomber sur la formule :
$$
\hat w= (X^T X) ^{-1} X^T Y
$$
Aide: Pour calculer la différentielle d'une fonction très simple à plusieurs variables, on peut simplement essayer de trouver une formule du type: 
$$
   f(w+\epsilon) = f(w) + \hbox{differential}(\epsilon) + o (\epsilon) 
$$

#  Des inputs plus nombreux


Les moindres carrés se généralise immédiatement avec $n$-input. Seule la représentation graphique est moins sympa. 

Observez le programme suivant. Essayer de vous représenter le plot obtenu en 3D. Eventuellement, tracez-le avec matplotlib en 3D. 

```python
def createData_two_output(nbData: int, sigma=0.3):
    x = np.random.random([nbData, 2]) * 2
    w = np.array([2., 4])
    w0 = 5
    '''y[i] = sum_j x[ij] w[j]  '''
    y = np.sum(x * w, axis=1) + w0 +  np.random.choice(a=[2,-2],p=[0.5,0.5],size=[nbData])
    return x, y

nbData=100
x,y=createData_two_output(nbData)
x_ext=np.zeros(shape=[nbData,3])
x_ext[:,0]=np.ones(shape=nbData)
x_ext[:,1:]=x

""" ( X^T X )^(-1) """
XTX_1 = np.linalg.inv(x_ext.T @ x_ext)
hat_w = XTX_1 @ x_ext.T @ y
#hat_Y= x_ext @ hat_w

print(hat_w)

plt.figure(figsize=(12,12))
mini,maxi=np.min(y),np.max(y)
xx=np.linspace(0,2)
x_1, x_2 = np.meshgrid(xx, xx)
hat_Y= hat_w[0]+hat_w[1]*x_1+hat_w[2]*x_2
plt.imshow(hat_Y,
           origin='lower',
           extent=[0,2,0,2],
           norm=plt.Normalize(vmin=mini, vmax=maxi))


plt.scatter(
            x[:, 0],
            x[:, 1],
            marker='o',
            c= y,
            edgecolors="k",
            #cmap=self.colorMap,
            norm=plt.Normalize(vmin=mini, vmax=maxi)
        )
plt.xlim(0,2)
plt.ylim(0,2)

plt.colorbar();
#plt.plot(x[:,0],x[:,1],'o',c=y)



```

# Le bruit est maintenant gaussien (ou pas loin)

Maintenant on suppose précisément que nos observations `y[i]` (petit y) sont des réalisations de va $Y[i]$ (grand Y) telles que: 
$$
    Y[i] = w .\mathtt x[i] + bruit[i]
$$
et que `bruit[i]` ce sont des v.a gaussiennes centrées de variance σ² (qui ne dépend pas de i). En d'autre terme:

    Y[i] ~ Normale (esp = μ[i] , var = σ² ),   μ[i] = w . x[i]

La densité de probabilité des `Y[i]` c'est, à une constante près: 
$$
      \prod_i \exp (\frac{ y_i - w x[i]}{2 \sigma}  )
$$
Donc le paramètre $w$ qui rend le plus vraisemblable les observations `y[i]` (petit y) c'est tout simplement
$$
  \hat w= \hbox{argmax}_i \prod_i \exp (\frac{ y_i - w \mathtt x[i]}{2 \sigma}  )
$$

*** Exo: *** Par une habile utilisation de la fonction log, montrez que l'on retombe exactement sur le $\hat w$ des moindres carrés ! 


De plus, dans ce modèle probabiliste très précise, on peut aussi estimer la variance du bruit, la variances des estimateurs $\hat w_i$, ce qui nous permettra aussi de savoir si l'un des $w_i$ est probablement nul. Ce qui revient à dire que le i-ième input n'a aucune influence sur l'output. Ce genre de considération est importante, car on chercher à espliquer l'output de la manière la plus simple possible.  


```python
""" observez la génération des données: parmi les input, deux n'ont aucune influence  """
def createData_four_input(nbData: int, sigma=0.3):
    b = 5
    W = np.array([0., 2, 1, 0])
    x = np.random.random([nbData, 4]) * 2
    y = np.sum(x * W, axis=1) + b + np.random.normal(0, sigma, size=[nbData])
    return x, y


nbData=100
nbInput=4
x,y=createData_four_input(nbData)
x_ext=np.zeros(shape=[nbData,nbInput+1])
x_ext[:,0]=np.ones(shape=nbData)
x_ext[:,1:]=x

""" ( X^T X )^(-1) """
XTX_1 = np.linalg.inv(x_ext.T @ x_ext)
w_hat = XTX_1 @ x_ext.T @ y
Y_hat= x_ext @ w_hat

""""    Estimateur de la variance du bruit
Au dénominateur, le ( - nbDescriptors -1)  c'est pour rendre l'estimateur sans biais   """
sigma_hat = np.sqrt(np.sum((y - Y_hat) ** 2) / (nbData - nbInput - 1))
print("sigma_hat: ",sigma_hat)


"""  Estimateur de la variance des w_hat
(c'est la variance du bruit multipliée par un coefficient qui dépend de la matrice X"""
std_err= sigma_hat * np.sqrt(np.diag(XTX_1))


"""calcul de la statistique signée  (T-score).
* T_score[i] petit  =>  hypothèse H_0[i]  :  w[i] est nulle (il ne sert à rien dans le modèle)
* T_score[i] grand  =>  hypothèse H_1[i]  :  w[i] est non-nulle"""
T_scores= w_hat / std_err

"""p-value associée
*p-value[i]  petite => hypothèse H_1[i]  => on garde le i-ème descripteur
*p-value[i]  grande => hypothèse H_0[i]  => on supprime le i-ème descripteur """
pValues=2*stats.t.sf(np.abs(T_scores),nbData-nbInput)

print("pValues: ",pValues)

```
